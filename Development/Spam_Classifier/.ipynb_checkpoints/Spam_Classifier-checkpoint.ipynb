# Data Inspection

```python
# Label distribution
spam_dist = len(df[df['label'] =='spam'])
ham_dist = len(df[df['label'] =='ham'])
print('Spam distribution: ', spam_dist)
print('Ham distribution: ', ham_dist)

# Check for missing values
print("Label Nulls: {}".format(df['label'].isnull().sum()))
print("Message Nulls: {}".format(df['message'].isnull().sum()))
```

```python
stopword = nltk.corpus.stopwords.words('english')

def clean_text(text):
    raw = "".join([word for word in text if word not in string.punctuation]) #remove punctuation
    tokens = re.split("\W+", raw) #tokenize
    text = [word for word in tokens if word not in stopword] #remove stopwords
    return text

df['text_clean'] = df['message'].apply(lambda x: clean_text(x.lower()))
df.head()
```

# Stemming

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
dir(ps)
```

```python
def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

df['text_stemmed'] = df['text_clean'].apply(lambda x: stemming(x))
df.head()
```

# Lemmatizing

```python
from nltk.stem import WordNetLemmatizer
wn1 = WordNetLemmatizer()
print(wn1.lemmatize('dogs'))
```

```python
def lemmatizing(text):
    text = [wn1.lemmatize(word) for word in text]
    return text

df['text_lemmatized'] = df['text_clean'].apply(lambda x: lemmatizing(x))
df.head()
```

# Vectorizing

```python
# Count Vectorization
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(analyzer = lemmatizing)
X_counts = count_vect.fit_transform(df['text_lemmatized'])
print(X_counts.shape)
print(count_vect.get_feature_names())#print unique words in messages
```

```python
data_sample = df[0:20]
count_vect_sample = CountVectorizer(analyzer = lemmatizing)
X_counts_sample = count_vect_sample.fit_transform(data_sample['text_lemmatized'])
print(X_counts_sample.shape)
print(count_vect_sample.get_feature_names())
```

```python
#Store non-zero values - create document term matrix
X_counts_df = pd.DataFrame(X_counts_sample.toarray())
X_counts_df
```

# Vectorizing Raw Data: N-Grams

```python
#dataframe column width
pd.set_option('display.max_colwidth', 100)

#Read in spam dataset containing classification and messages
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])
print(df)
```

```python
stopword = nltk.corpus.stopwords.words('english')

def clean_text(text):
    raw = "".join([word for word in text if word not in string.punctuation]) #remove punctuation
    tokens = re.split("\W+", raw) #tokenize
    text = " ".join([ps.stem(word) for word in tokens if word not in stopword]) #remove stopwords
    return text

df['text_clean'] = df['message'].apply(lambda x: clean_text(x.lower()))
df.head()
```

```python
#Apply CountVectorizer (w/ N-Grams)
from sklearn.feature_extraction.text import CountVectorizer
ngram_vect = CountVectorizer(ngram_range=(2,2)) #unigrams, 1,2 unigrams & bigrams, 1,3 uni, bi and trigrams 
X_counts = ngram_vect.fit_transform(df['text_clean'])
print(X_counts.shape)
print(ngram_vect.get_feature_names())
```

```python
#Apply CountVectorizer (w/ N-Grams) to a smaller sample
data_sample = df[0:20]

ngram_vect_sample = CountVectorizer(ngram_range=(2,2)) #unigrams, 1,2 unigrams & bigrams, 1,3 uni, bi and trigrams 
X_counts_sample = ngram_vect_sample.fit_transform(data_sample['text_clean'])
print(X_counts_sample.shape)
print(ngram_vect_sample.get_feature_names())
```

```python
#convert sparse matrix to dataframe
X_counts_df = pd.DataFrame(X_counts_sample.toarray())
X_counts_df.columns = ngram_vect_sample.get_feature_names()
X_counts_df
```

# Feature Engineering

```python
#dataframe column width
pd.set_option('display.max_colwidth', 100)

#Read in spam dataset containing classification and messages
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])
print(df)
```

```python
#Create feature for text message length
df['body_len'] = df['message'].apply(lambda x: len(x) - x.count(" "))
df.head()
```

```python
#Create feature for % of text that is punctuation
import string

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")), 3)*100

df['punct%'] = df['message'].apply(lambda x: count_punct(x))
df.head()
```

```python
#Evalute created features to determine if the two previously created features are valuable
from matplotlib import pyplot 
import numpy as np
%matplotlib inline
```

```python
bins = np.linspace(0, 200, 40) #define bins with array with 40 evenly spaced numbers between 0-200
# alpha: how dark 
# normed: normalize to the same scale 

pyplot.hist(df[df['label']=='spam']['body_len'], bins, alpha=0.5, density=True, label='spam') #filter data to see only rows where spam, look at body length column for rows
pyplot.hist(df[df['label']=='ham']['body_len'], bins, alpha=0.5, density=True, label='ham')
pyplot.legend(loc='upper left')
pyplot.show()
#Spam messages look to be longer than ham messages
```

```python
# Check punctuation feature value
bins = np.linspace(0, 50, 40) #define bins with array with 40 evenly spaced numbers between 0-200
# alpha: how dark 
# normed: normalize to the same scale 

pyplot.hist(df[df['label']=='spam']['punct%'], bins, alpha=0.5, density=True, label='spam') #filter data to see only rows where spam, look at body length column for rows
pyplot.hist(df[df['label']=='ham']['punct%'], bins, alpha=0.5, density=True, label='ham')
pyplot.legend(loc='upper right')
pyplot.show()
#Spam shows slightly higher punctuation usage but not enough to validate that this feature is valuable
```

```python
#Evaluation Metrics: Accuracy, Precision, Recall
import nltk
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
import string
from nltk.stem import PorterStemmer

stopword = nltk.corpus.stopwords.words('english')
ps = PorterStemmer()

#dataframe column width
pd.set_option('display.max_colwidth', 100)

#Read in spam dataset containing classification and messages
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])
print(df)

def count_punct(text):
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")), 3)*100

df['body_len'] = df['message'].apply(lambda x: len(x) - x.count(" "))
df['punct%'] = df['message'].apply(lambda x: count_punct(x))
df.head()
```

```python
#Ensemble Method - multiple models combined to be better than each model individual models.
#Random Forest - Collection of Decision Trees combined to strengthen overall Ensemble model. 



def clean_text(text):
    raw = "".join([word for word in text if word not in string.punctuation]) #remove punctuation
    tokens = re.split("\W+", raw) #tokenize
    text = [ps.stem(word) for word in tokens if word not in stopword] #remove stopwords
    return text

df['text_clean'] = df['message'].apply(lambda x: clean_text(x.lower()))
df.head()

tfidf_vect = TfidfVectorizer(analyzer=clean_text)
X_tfidf = tfidf_vect.fit_transform(df['text_clean'])

X_features = pd.concat([df['body_len'], df['punct%'], pd.DataFrame(X_tfidf.toarray())], axis=1)
X_features.head()

```

# Model Training

```python
#define model training function using supervised learning methods based on applying Bayes' theorem with strong (naive) feature independence assumptions.
def train_model(classifier, feature_vector_train, label, feature_vector_test, is_neural_net=False):
    #fit training dataset to classifier
    classifier.fit(feature_vector_train, label)
    #predict labels on the test dataset
    predictions = classifier.predict(feature_vector_test)
    
    return metrics.accuracy_score(predictions, y_test)
```

# Model Training - Naive Bayes

```python
NBaccuracy = train_model(naive_bayes.MultinomialNB(alpha=0.2), x_train_tfidf, y_train, x_test_tfidf)
print("Naive Bayes Accuracy: ", NBaccuracy)
```
